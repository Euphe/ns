```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
plt.style.use('ggplot')
```

```python
def randgp(n, p):
    G = nx.Graph()
    G.add_nodes_from(range(0,n))
    
    for i in xrange(0,n):
        for j in xrange(0,n):
            if rnd.random() <= p:
                G.add_edge(i,j)
        
    return G

def randgm(n, m):
    G = nx.Graph()
    G.add_nodes_from(range(0,n))
    edges = []
    for i in xrange(0,n):
        for j in xrange(0,n):
            edges.append((i,j))
    
    num = n**2
    for i in xrange(0,m):
        e = int(np.floor(rnd.random()*num))
        edge = edges.pop(e)
        G.add_edge(edge[0],edge[1])
        num -= 1
    
    return G
```

### Clustering coefficient

During the lecture we have understood, that the clustering coefficient of a random graph is equal to the probability $p$: $$\text{Clustering coefficient} = \frac{\langle k \rangle}{n} = p $$

In this task you have to check it on generated data.
Please, generate $100$ Random Graphs with $n = 1000$ and $p = 0.002$ (for saving computational time) and plot the box-plot of your computations.

```python
num_graphs = 100
n = 1000
p = 0.002

graphs = [nx.gnp_random_graph(n, p) for i in range(num_graphs)]
graph_clustering_coefficients = np.array([np.mean(g.degree().values())/n for g in graphs])
```

```python
plt.boxplot(graph_clustering_coefficients)
plt.show()
```

Evidently most values fall between $0.00205$ and $0.00195$, with the mean being $0.002 = p$. 

### Size of small components

In this task you are asked to calculate the average size of small components (small component = not a giant one) with regard to average degree of the network. To see the effect clearly, plot average size around $\langle k \rangle = 1$.

```python
num_graphs = 10
n = 1000
p_initial = 0.0001
p_end = 0.003
p_values = np.linspace(p_initial, p_end, 20)

graphs = []
for p in p_values:
    graphs += [nx.gnp_random_graph(n, p) for i in range(num_graphs)]
```

```python
average_degrees = []
average_component_sizes = []
for g in graphs:
    mean_degree = np.mean(g.degree().values())
    average_degrees.append(mean_degree)
    components = list(nx.connected_component_subgraphs(g))[1:]
    component_sizes = [component.size() for component in components]
    mean_component_size = np.mean(component_sizes)
    average_component_sizes.append(mean_component_size)
```

```python
plt.scatter(average_degrees, average_component_sizes)
plt.xlabel('Average node degree')
plt.ylabel('Average small component size')
plt.show()
```

This is not very descriptive. Most components are of size 1, so the average small component size tends to gravitate to $1$ too much. This is due to small $p$ values considered.

Let's try to omit the $5$ biggest components of each graph.

```python
average_degrees = []
average_component_sizes = []
for g in graphs:
    mean_degree = np.mean(g.degree().values())
    average_degrees.append(mean_degree)
    components = list(nx.connected_component_subgraphs(g))[5:]
    component_sizes = [component.size() for component in components]
    mean_component_size = np.mean(component_sizes)
    average_component_sizes.append(mean_component_size)
```

```python
plt.scatter(average_degrees, average_component_sizes)
plt.xlabel('Average node degree')
plt.ylabel('Average small component size')
plt.show()
```

Now we can see a critical point at $k=1$ after which the average small component size starts to drop.

### Average path length and transitivity

Plot the average path length and transitivity coefficents with $n=1000$ and p=0.001 while increasing by 0.003 

```python
num_graphs = 10
n = 50
p_initial = 0.001
p_values = np.linspace(p_initial, 1, 30)

avg_path_lens = []
avg_trans_coefs = []
for p in p_values:
    graphs = [nx.gnp_random_graph(n, p) for i in range(num_graphs)]
    local_avg_path_lens = []
    local_avg_trans_coefs = []
    
    for graph in graphs:
        comp = next(nx.connected_component_subgraphs(graph))
        try:
            local_avg_path_lens.append(nx.average_shortest_path_length(comp))
        except ZeroDivisionError:
            continue
        local_avg_trans_coefs.append(nx.transitivity(comp))
    
    avg_path_lens.append(np.mean(local_avg_path_lens))
    avg_trans_coefs.append(np.mean(local_avg_trans_coefs))
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True) 
ax1.plot(p_values, avg_path_lens)
ax1.set_title('Average shortest path lengths')
ax2.plot(p_values, avg_trans_coefs)
ax2.set_title('Average transitivity coefficients')
plt.xlabel('p')
plt.show()
```

We take 30 points (instead of 300) via `np.linspace` due to slow computation.

Some graphs, for $p < 0.02$, don't contain any edges. Average shortest paths have not been computed for these graphs.

From the plots a critical value of $p = 0.01$ is evident: after it there is a rapid drop in average shortest path lengths. It's the point after which the graphs become very interconnected.
