b) In order to see how fast it converges, we shall need to code our own Page Rank algorithm. Code a function ```mypagerank(G,alpha,k)``` which executes the power iteration $k$ times starting from the uniform distribution among nodes in the graph. It should return the list of page rank scores. Recall that
$$ \mathbf{p}^{t+1} = \alpha (D^{-1}A)^{\top}\mathbf{p} + (1-\alpha)\frac{\mathbf{e}}{n}, $$
where $A$ is the adjacency matrix of $G$ and $D$ is the diagonal matrix made of degrees of vertices.

````python
## Reminder: A @ B computes the multiplication of matrices A and B.

def mypagerank(graph,alpha,k):
    adjacency_matrix = nx.adjacency_matrix(graph).todense()
    node_amount = len(adjacency_matrix)
    diags = np.array(adjacency_matrix.sum(axis=0)).flatten()
    D = np.diag(diags)
    p = (1/node_amount * np.ones((adjacency_matrix.shape[0],))).reshape((node_amount, 1))
    for i in range(k):
        p = alpha * (np.linalg.inv(D) @ adjacency_matrix).T @ p + (1 - alpha) * np.ones((adjacency_matrix.shape[0],1))/node_amount
    
    return np.array(p).flatten()

mypagerank(G, 0.1, 1)
````

Alright, let us see how it goes...

c) Plot the results after 0 up to 10 iterations.

````python
fig = plt.figure(1, figsize=(7,7))
ax = plt.subplot(111)    

its = np.arange(0,4, 1)

for it in its:
    pr = mypagerank(G, 0.85, it)
    prval = list(pr)
    ax.plot(prval, label='{:d} iteration(s)'.format(it))#, alpha = it/len(its)+0.2)
    
ax.legend()
````

d) Pay attention to the order of colors. What do you observe? How can you explain it?

Initially we have uniform pageranks, hence the horizontal line.

After one iteration the highly connected nodes get a lot of pagerank, as all nodes provide equal pagerank. The weakly connected nodes get less pagerank. So we get the jagged orange line.

After two iterations the line "jumps back" towards the horizontal line. Highly connected nodes get less pagerank, weakly connected nodes get more pagerank.

After three iterations the highly connected nodes get more pagerank, and the weakly connected ones get less pagerank **again**.

We draw a conclusion that the algorithm does not converge monotonically. It tends to extend and retract.

The horizontal line divides the highly connected and the weakly connected nodes.

Experiment several teleportation vectors on the coappearance network of characters in the novel _Les Misérables_ (V. Hugo). The graph is in the file `lesmis.gml`. It was compiled by Donald Knuth [1]. Try to focus on several famous characters (Valjean, Javert, Gavroche, Cosette), or use betweenness centrality (is degree centrality of interest here?) for the teleportation. Each time, draw the network induced by nodes with high page rank. What can you observe?


[1] D. E. Knuth, _The Stanford GraphBase: A Platform for Combinatorial Computing_, Addison-Wesley, Reading, MA (1993).

````python
graph = nx.read_gml('lesmis.gml')
pos = nx.spring_layout(graph)

nx.draw_networkx(graph,pos)
````

````python
def draw(G, pos, sizes, colors, title, size_mult=50000):
    nodes = nx.draw_networkx_nodes(G, pos,
                                   cmap=plt.cm.copper, 
                                   node_size=[c*size_mult for c in sizes],
                                   node_color=colors,
                                   nodelist=G.nodes())
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1))
    
    pos_higher = {}
    y_off = 0.05

    for k, v in pos.items():
        pos_higher[k] = (v[0], v[1]+y_off)

    labels = nx.draw_networkx_labels(G, pos_higher)
    edges = nx.draw_networkx_edges(G, pos, alpha=0.2)
    
    plt.title(title)
    plt.axis('off')
    plt.show()
````

````python
plt.figure(1, figsize=(20,10))
eye_vector = np.ones((len(graph.nodes()))) * 1/len(graph.nodes())
zeros_vector = np.zeros((len(graph.nodes())))
draw(graph, pos, eye_vector, eye_vector, 'huh')
````

````python
# Let's try degree centrality
pos = nx.spring_layout(graph)
dc = nx.degree_centrality(graph)
degree_centralities = np.array(list(dc.values()))
plt.figure(1, figsize=(20,10))
draw(graph, pos, eye_vector, degree_centralities, 'Degree centrality')
````

````python
# Let's try betweennes centrality
pos = nx.spring_layout(graph)
dc = nx.betweenness_centrality(graph)
betweenness_centrality = np.array(list(dc.values()))
plt.figure(1, figsize=(20,10))
draw(graph, pos, eye_vector, degree_centralities, 'Betweenness centrality')
````

````python
top_vals_amount = 20
````

````python
# Extract a subgraph by top betweenness centrality
dcs = pd.Series(betweenness_centrality)
nodes = pd.Series(graph.nodes())
top = nodes[np.argsort(dcs)][-top_vals_amount:]
mask = top.index
nodes.index = dcs.index
subgraph_nodes = nodes.ix[mask]
subgraph_dcs = dcs.ix[mask]
subgraph = nx.subgraph(graph, subgraph_nodes)
plt.figure(1, figsize=(20,10))
draw(subgraph, nx.spring_layout(subgraph), eye_vector, subgraph_dcs, 'Betweenness centrality top %d vertices' % (top_vals_amount))
````

````python
print(pd.DataFrame(pd.concat([subgraph_nodes, subgraph_dcs], axis=1)))
````

````python
# Raw pagerank
pranks_dict = nx.pagerank(graph)
pranks = np.array(list(pranks_dict.values()))
plt.figure(1, figsize=(20,10))
draw(graph, pos, eye_vector, pranks, 'Pagerank')
````

````python
def draw_top(graph, centralities, top_vals_amount=20, title=''):
    # Extract a subgraph by top betweenness centrality
    dcs = pd.Series(centralities)
    nodes = pd.Series(graph.nodes())    
    top = nodes[np.argsort(dcs)][-top_vals_amount:]
    mask = top.index
    nodes.index = dcs.index
    subgraph_nodes = nodes.loc[mask]
    subgraph_dcs = dcs.loc[mask]
    subgraph = nx.subgraph(graph, subgraph_nodes)
    plt.figure(1, figsize=(20,10))
    normed_subgraph_dcs = subgraph_dcs/subgraph_dcs.sum()
    draw(subgraph, nx.spring_layout(subgraph), eye_vector, normed_subgraph_dcs, '%s top %d vertices' % (title, top_vals_amount))
plt.figure(1, figsize=(20,10))
draw_top(graph, pranks, title='Pagerank')
````

````python
# Main characters = Valjean, Javert, Gavroche, Cosette

valjean_index = list(graph.nodes()).index('Valjean')
javert_index = list(graph.nodes()).index('Javert')
gavroche_index = list(graph.nodes()).index('Gavroche')
cosette_index = list(graph.nodes()).index('Cosette')
````

````python
# Pagerank with teleporting to Valjean
valjean_personalization = {k: 0 for k in graph.nodes()}
valjean_personalization['Valjean'] = 1

pranks_dict = nx.pagerank(graph, personalization=valjean_personalization)
pranks = np.array(list(pranks_dict.values()))
plt.figure(1, figsize=(20,10))
draw_top(graph, pranks, title='Pagerank Valjean personaloization')
````

We can see two communities around Valjean. These are different chacters of the novel that interacted with Valjean.

````python
# Pagerank with teleporting to Gavroche
gavroche_personalization = {k: 0 for k in graph.nodes()}
gavroche_personalization['Gavroche'] = 1

pranks_dict = nx.pagerank(graph, personalization=gavroche_personalization)
pranks = np.array(list(pranks_dict.values()))
plt.figure(1, figsize=(20,10))
draw_top(graph, pranks, title='Pagerank Gavroche personaloization')
````

## Task 3

Let us now focus on larger networks. First let us read the network of political blogs.

````python
#Some arcs are multi and it prevents some algorithms to work. We thus simplify it.

G = nx.read_gml('polblogs.gml')
A = nx.adjacency_matrix(G)
M = A.astype(bool).astype(int) #converts non-zero entries to True and then True to 1.
H = nx.DiGraph(M)
L = list(G)
mapping={}
for i in range(len(L)):
    mapping[i]=L[i]
Gsimp=nx.relabel_nodes(H,mapping) #relabel vertices with original names.
````

Run the HITS algorithm on `Gsimp` and plot the hubs score against the authorities score.

````python
(h, a) = nx.hits(Gsimp)
````

````python
la = list(a.values())
lh = list(h.values())
plt.plot(lh,la, '+')
plt.xlabel('Hubs')
plt.ylabel('Auth')
````

Vizualize top Authorities colored in blue with sizes proportional to A-value and their common Hubs coloured in green with size proportional to H-value.

````python
# Task:
# Pick the top biggest authorities
# Draw them with size proportional to A score
# And pick hubs
# Criteria:
# 1. Hub has high hub score and is connected to selected authorities
# Draw hubs in green with color depth proportional to H-value
````

````python
top_vals_amount = 10
````

````python
nodes = pd.Series(Gsimp.nodes())

a_vals = pd.Series(list(a.values()))
h_vals = pd.Series(list(h.values()))

top_a = nodes[np.argsort(a_vals)][-top_vals_amount:]
top_h = nodes[np.argsort(h_vals)][-top_vals_amount:]
````

````python
# Authorities subgraph
plt.figure(1, figsize=(20,10))
draw_top(Gsimp, a_vals, title='Authorities graph', top_vals_amount=top_vals_amount)
````

````python
# Hubs subgraph
plt.figure(1, figsize=(20,10))
draw_top(Gsimp, h_vals, title='Hubs graph', top_vals_amount=top_vals_amount)
````

````python
# get top hubs
good_hubs = []
for hub in top_h:
    for authority in top_a:
        if Gsimp.has_edge(hub, authority) or Gsimp.has_edge(authority, hub):
            good_hubs.append(hub)
            break
len(good_hubs)
# All top hub value hubs are connected to at least one top authority
````

````python
# Plot hubs and authorities together
authority_nodes = list(top_a.values)
hub_nodes =  list(top_h.values)
nodes = authority_nodes + hub_nodes

normed_avals = a_vals[np.argsort(a_vals)][-top_vals_amount:]
normed_avals = list(normed_avals*20/normed_avals.sum())
normed_hvals = list(h_vals[np.argsort(h_vals)][-top_vals_amount:])
normed_hvals = list(np.array(normed_hvals)/np.sum(normed_hvals))

authority_sizes = list(normed_avals) 
hub_sizes = [0.5 for i in range(len(top_h))]
authority_colors = ['black' for i in range(len(top_a))]
hub_colors = normed_hvals

subgraph = nx.subgraph(Gsimp, nodes)
````

````python
plt.figure(figsize=(30,20))

size_mult = 4000


pos = nx.spring_layout(subgraph)

nodes_draw = nx.draw_networkx_nodes(subgraph, pos,
                               node_size=[c*size_mult for c in authority_sizes],
                               node_color=authority_colors,
                               nodelist=authority_nodes)
nodes_draw.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1))
nodes_draw = nx.draw_networkx_nodes(subgraph, pos,
                               node_size=[c*size_mult for c in hub_sizes],
                               cmap = plt.cm.Greens, 
                               node_color=hub_colors,
                               nodelist=hub_nodes)
nodes_draw.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1))

pos_higher = {}
y_off = 0.1

for k, v in pos.items():
    pos_higher[k] = (v[0], v[1]+y_off)

labels = nx.draw_networkx_labels(subgraph, pos_higher)
edges = nx.draw_networkx_edges(subgraph, pos, alpha=0.1)

plt.title('Polblogs hubs and authorities')
plt.axis('off')
plt.show()
````

Here we can see authorities - black nodes. Node size is proportional to A-score.
Green nodes are hubs. Color depth is proportional to H-score.

Curiously one node is both a strong hub and a large authority.
